```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/netflix_titles.csv")
```

```python
df.head(10)
```

```python
df.dropna(inplace = True)
df.reset_index(inplace = True)
df
```

```python
newdf  = len(df[(df['release_year']==2021)&(df['type']=="Movie")])
```

```python
newdf
```

```python
avengers_df = df[df['title'].str.contains('Avengers')]
avengers_df
```

```python
smol_df= df['country'].value_counts()
smol_df
```

```python
top = smol_df.head(5)
top
```

```python
result_df = df[(df['country'].isin(top.index))]

result_df.reset_index(inplace = True)
result_df
```

```python

plt.pie(top.values, labels=top.index, autopct='%1.1f%%')
plt.title('No.of movies per Country')
plt.show()
```
